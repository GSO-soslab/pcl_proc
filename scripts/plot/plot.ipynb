```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
file2 = pd.read_csv("/home/soslab/auv_ws/test4.csv")
```

```python
column_values = file2['0.0'].tolist()[:-50]
```

RMSE = √Σ(Pi – Oi)2 / n

```python
column_values = column_values
value = 0
for i in column_values:
    value += ((15 - i) ** 2)
```

```python
rmse = np.sqrt(value/len(column_values))
```

```python
plt.figure(figsize=(40, 20))
plt.plot(column_values, linewidth=6)
plt.annotate(xy=(601,20), text=f"RMSE: {round(rmse,4)}", fontsize=45)

plt.axhline(y=15, color='r', linestyle='--', linewidth=15)
plt.annotate(xy=(-20,15.5), text=f"Desired Standoff (15m)", color='r',fontsize=40)
plt.axhline(y=2, color='r', linestyle='--', linewidth=15)
plt.annotate(xy=(-20,2.5), text=f"Minimum Detectable Range by Sonar (2m)", color='r',fontsize=40)

plt.xlabel('''Time [s]''', fontsize = 90)
plt.ylabel('Standoff distance [m]', fontsize = 90)
plt.grid(True)
plt.xticks(fontsize=70)
plt.yticks(fontsize=70)
```

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
file = pd.read_csv("/home/soslab/auv_ws/test4_odom.csv")
```

```python
# x = file["16.90502493263138"].tolist()[:-200]
# y = file['12.78448097322282'].tolist()[:-200]

y = file['12.916642178284103'].tolist()
x = file["5.46589024861444"].tolist()
```

```python
y_trans = []
for i in y:
    i = -i
    y_trans.append(i)
```

```python
plt.xlabel('''y  [m]''', fontsize = 15)
plt.ylabel("x  [m]", fontsize = 15)

plt.scatter(y_trans[0],x[0],color="red",label='Vehicle Initial Pose')
plt.scatter(y_trans[-50],x[-50],color="#4eea60",label='Vehicle Final Pose')

plt.scatter(30,45,color="#d04eea",marker='^', label='Iceberg Initial Pose (30,45)')
plt.scatter(69,60,color="#1a46b6",marker='^',label='Iceberg Final Pose (69,60)')
plt.legend(loc=(1,0.5), fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.plot(y_trans[:-100],x[:-100], linewidth=3)
```
